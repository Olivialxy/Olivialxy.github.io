```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
df = pd.read_csv(url)
df.head()
```

```python
agency_sqft = df.groupby('Agency Name')['Square Footage'].sum().reset_index()

chart1 = alt.Chart(agency_sqft).mark_bar().encode(
    x=alt.X('Square Footage:Q', title='Total Square Footage'),
    y=alt.Y('Agency Name:N', sort='-x', title='Agency'),
    color=alt.Color('Square Footage:Q', scale=alt.Scale(scheme='blues'))
).properties(
    title='Total Square Footage by Illinois State Agency',
    width=700
)
chart1
```

```python
df.columns = df.columns.str.strip()

df = df.dropna(subset=['Usage Description'])

chart_data = df.groupby(['City', 'Usage Description']).size().reset_index(name='count')

usage_dropdown = alt.binding_select(options=sorted(chart_data['Usage Description'].unique()))
usage_select = alt.selection_single(fields=['Usage Description'], bind=usage_dropdown, name="Usage")

chart2 = alt.Chart(chart_data).mark_bar().encode(
    x=alt.X('count:Q', title='Number of Buildings'),
    y=alt.Y('City:N', sort='-x'),
    color=alt.Color('City:N'),
    tooltip=['City:N', 'count:Q']
).add_params(
    usage_select
).transform_filter(
    usage_select
).properties(
    width=700,
    height=400,
    title='Number of Buildings per City by Usage Type (Interactive)'
)

chart2
```
